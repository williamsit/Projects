```python
import os
import csv
import pandas as pd
```

```python
religion_loc = os.path.join("Resources", "religion.csv")

with open(religion_loc, newline="") as religion_csv:
    religion_reader = csv.reader(religion_csv, delimiter=",")
    print(religion_reader)
    
uv_loc = os.path.join("Resources", "uv.csv")

with open(uv_loc, newline="") as uv_csv:
    uv_reader = csv.reader(uv_csv, delimiter=",")
    print(uv_reader)
```

```python
religion_df = pd.read_csv(os.path.join("Resources", "religion.csv"))
religion_df.head()
```

```python
uv_df = pd.read_csv(os.path.join("Resources", "uv.csv"))
uv_df.head()
```

```python
religion_df.columns
```

```python
religion_renamed_df = religion_df.rename(columns={"fips":"FIPS Code", 
                                                  "stcode":"State Code",
                                                 "stabbr":"State Abbreviation",
                                                 "stname":"State Name", 
                                                 "cntycode":"County Code",
                                                 "cntyname":"County Name",
                                                 "pop2010":"Population in 2010", 
                                                 "totcng":"Total Number of Congregations",
                                                 "totadh":"Total Number of Adherents",
                                                 "totrate":"Rates of Adherence per 1,000 population",
                                                 "evancng":"Evangelical Protestant Congregations", 
                                                 "evanadh":"Evangelical Adherents",
                                                 "evanrate":"Evangelical Rate of Adherence per 1,000 population",
                                                 "bprtcng":"Black Protestant Congregations",
                                                 "bprtadh":"Black Protestant Adherents",
                                                 "bprtrate":"Black Protestant Rate of Adherence per 1,000 population",
                                                 "mprtcng":"Mainline Protestant Congregations",
                                                 "mprtadh":"Mainline Protestant Adherents",
                                                 "mprtrate":"Mainline Protestant Rate of Adherence per 1,000 population",
                                                 "cathcng":"Catholic Congregations",
                                                 "cathadh":"Catholic Adherents",
                                                 "cathrate":"Catholic Rate of Adherence per 1,000 population"})
religion_renamed_df.columns
```

```python
uv_df.columns
```

```python
uv_renamed_df = uv_df.rename(columns={"STATENAME":"State Name",
                                     "STATE_ABR":"State Abbreviation",
                                     "COUNTY_FIPS":"FIPS Code",
                                     "COUNTY NAME":"County Name",
                                     "UV_ Wh/m_":"UV Exposure in Meters Squared"})
uv_renamed_df.head()
```

```python
religion_renamed_df.head()
```

```python
religion_renamed_dropped_df = religion_renamed_df.dropna(how="any")
religion_renamed_dropped_df.count()
```

```python
uv_renamed_dropped_df = uv_renamed_df.dropna(how="any")
uv_renamed_dropped_df.count()
```

```python
uv_renamed_dropped_df.head()
```

```python
religion_renamed_dropped_df.head()
```

```python
religion_county_column = religion_renamed_dropped_df["County Name"]
religion_county_new_column = []
for each_county in religion_county_column:
    each_county_list = each_county.split(" ")
    if "County" in each_county_list:
        each_county_list.remove("County")
        only_county_name = " ".join(each_county_list)
        religion_county_new_column.append(only_county_name)
    elif "Parish" in each_county_list:
        each_county_list.remove("Parish")
        only_county_name = " ".join(each_county_list)
        religion_county_new_column.append(only_county_name)
    elif "city" in each_county_list:
        each_county_list.remove("city")
        only_county_name = " ".join(each_county_list)
        religion_county_new_column.append(only_county_name)
    elif "Municipality" in each_county_list:
        each_county_list.remove("Municipality")
        only_county_name = " ".join(each_county_list)
        religion_county_new_column.append(only_county_name)
    elif "Borough" in each_county_list:
        each_county_list.remove("Borough")
        only_county_name = " ".join(each_county_list)
        religion_county_new_column.append(only_county_name)
    elif "Area" in each_county_list:
        each_county_list.remove("Area")
        only_county_name = " ".join(each_county_list)
        religion_county_new_column.append(only_county_name)
    elif "District" in each_county_list:
        only_county_name = " ".join(each_county_list)
        religion_county_new_column.append(only_county_name)
    else:
        print(each_county_list)
```

```python
religion_renamed_dropped_df["County Name"] = religion_county_new_column
religion_renamed_dropped_df.head()
```

```python
merged_df = pd.merge(religion_renamed_dropped_df, uv_renamed_df, on="FIPS Code")
merged_df.head()
```

```python
merged_renamed_df = merged_df.rename(columns={"State Abbreviation_x":"State Abbreviation",
                                    "State Name_x":"State Name", "County Name_x":"County Name"})
merged_renamed_df.head()
```

```python
merged_renamed_dropped_df = merged_renamed_df.drop(["State Name_y", "State Abbreviation_y", "County Name_y"], axis=1)
merged_renamed_dropped_df.head()
```

```python
merged_renamed_dropped_df_columns = list(merged_renamed_dropped_df.columns.values)
merged_renamed_dropped_df_columns
```

```python
merged_reordered_df = merged_renamed_dropped_df[['FIPS Code',
 'State Code',
 'County Code',
 'State Abbreviation',
 'State Name',
 'County Name',
 'Population in 2010',
 'Total Number of Congregations',
 'Total Number of Adherents',
 'Rates of Adherence per 1,000 population',
 'Evangelical Protestant Congregations',
 'Evangelical Adherents',
 'Evangelical Rate of Adherence per 1,000 population',
 'Black Protestant Congregations',
 'Black Protestant Adherents',
 'Black Protestant Rate of Adherence per 1,000 population',
 'Mainline Protestant Congregations',
 'Mainline Protestant Adherents',
 'Mainline Protestant Rate of Adherence per 1,000 population',
 'Catholic Congregations',
 'Catholic Adherents',
 'Catholic Rate of Adherence per 1,000 population',
 'UV Exposure in Meters Squared']]
merged_reordered_df.head()
```

```python
merged_reordered_df.to_csv(os.path.join("Resources", "merged_data.csv"),header=True,index=False)
```
